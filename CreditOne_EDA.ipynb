## EXPLORATORY DATA ANALYSIS

## Dataset Overview 

>The Credit Card Default payment in Taiwan  is used for exploratory data analysis to understand the credit card holders  behavior.

#### The attributes in the dataset are described as below 
    
* LIMIT_BAL: Amount of given credit in NT dollars (includes individual and family/supplementary credit
* SEX: Gender 
  * 1=male
  * 2=female
* EDUCATION: 
  * 1=graduate school, 
  * 2=university, 
  * 3=high school, 
  * 4=others
  * 5/6 = Unknown
* MARRIAGE:
  * 1=married, 
  * 2=single,
  * 3=others
* AGE: Age in years
* PAY_0,2-6: Repayment status September-April, 2005 
  * -1=pay duly, 
  * 1=payment delay for one month, 
  * 2=payment delay for two months....
  * 8=payment delay for eight months, 
  * 9=payment delay for nine months and above)
 * BILL_AMT 1-6:Amount of bill statement September-April, 2005 
 * PAY_AMT1 1-6: Amount of previous payment September-April, 2005 
 * default.payment.next.month: Default payment 
   * 1=yes(Default)
   * 0=no (Non-Default)
